# Working with Images

```python
# load and show an image with Pillow
from PIL import Image
# load the image
image = Image.open('/content/Sydney-Opera-House.jpg')
# show the image
image
```

```python
# summarize some details about the image
print(image.format)
print(image.mode)
print(image.size)
```

```python
# Convert the mode of image
# 'L' refers to the Greyscale mode
image.convert('L')
```

###Dithering Example
800px-Dithering_example_red_blue.svg.png

```python
# Convert the mode of image
# 'P' for dithering mode
# Dithering is creating the illusion of color depth in images on systems with limited color palettes
# Dithering introduces noise into an image and often can be visible
image.convert('P')
```

```python
# Convert the mode of image
# "1" is for bilevel mode, any RGB value bigger than 127 will be ceiled to 255 (representing the color white) and any RGB value less than 127 will be floored to 0 (representing the color black)
image.convert('1')
```

```python
# Crop image
# Coordinates are (left, upper, right, lower)
# (0, 0)-->the upper left corner

box = (100,100,300,300)
region = image.crop(box)
region
```

```python
# Paste image
region = region.convert('L')
image.paste(region,box)
image
```

# Working with Textual Data

```python
import pandas as pd
import numpy as np

# text creation
s = pd.Series(['Tom', 'William Rick', 'John',
               'Alber@t', np.nan, '1234',
               'SteveSmith'])
s
```

```python
#lower case text
s.str.lower()
```

```python
#upper case text
s.str.upper()
```

```python
#text concatenation
s.str.cat(sep='$')
```

```python
s.str.isnumeric()
```

```python
# Read text data to a dataframe
time_sentences = ["Monday: The doctor's appointment is at 2:45pm.",
                  "Tuesday: The dentist's appointment is at 11:30 am.",
                  "Wednesday: At 7:00pm, there is a basketball game!",
                  "Thursday: Be back home by 11:15 pm at the latest.",
                  "Friday: Take the train at 08:10 am, arrive at 09:00am."]

df = pd.DataFrame(time_sentences, columns=['text'])
df
```

```python
df['text'].str.findall('am')
```

```python
# split the text into tokens - splitting the string sentence by the spaces (essentially making it a list of string words)
splits = df['text'].str.split()
splits
```

```python
# the number of tokens for each string in df['text']
splits.str.len()
```

```python
# find which entries contain the word 'appointment'
df['text'].str.contains('appointment')
```

```python
df['text'].str.replace('at','AT')
```
